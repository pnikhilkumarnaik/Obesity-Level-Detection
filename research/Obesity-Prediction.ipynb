```python
from imblearn.over_sampling import SMOTE
from sklearn.utils import shuffle
```

```python
X = df.drop(columns=['NObeyesdad'])
y = df['NObeyesdad']
smote = SMOTE(sampling_strategy={0.0: 351, 1.0: 351, 2.0: 351, 3.0: 351, 4.0: 351, 5.0:351, 6.0:351})
X_resampled, y_resampled = smote.fit_resample(X, y)
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['NObeyesdad'])], axis=1)
df_resampled = shuffle(df_resampled, random_state=42).reset_index(drop=True)
df_resampled = df_resampled.sample(n=1000, random_state=42).reset_index(drop=True)
print(df_resampled['NObeyesdad'].value_counts())
print(f"Total records: {len(df_resampled)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import ExtraTreesClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
import time
```

```python
X=df.drop(columns=['NObeyesdad'])
Y=df['NObeyesdad']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=26)
```

```python
len(x_train),len(x_test)
```

```python
lr1=LogisticRegression()
start_time=time.time()
lr1.fit(x_train,y_train)
end_time=time.time()
lrt=end_time-start_time
lr1_pred=lr1.predict(x_test)
print("Accuracy of Logistic Regression:",accuracy_score(lr1_pred,y_test))
print(classification_report(lr1_pred,y_test))
```

```python
dt1=DecisionTreeClassifier()
start_time=time.time()
dt1.fit(x_train,y_train)
end_time=time.time()
dtt=end_time-start_time
dt1_pred=dt1.predict(x_test)
print("Accuracy of Decision Tree:",accuracy_score(dt1_pred,y_test))
print(classification_report(dt1_pred,y_test))
```

```python
rf1=RandomForestClassifier(n_estimators=500)
start_time=time.time()
rf1.fit(x_train,y_train)
end_time=time.time()
rft=end_time-start_time
rf1_pred=rf1.predict(x_test)
print("Accuracy of Random Forest:",accuracy_score(rf1_pred,y_test))
print(classification_report(rf1_pred,y_test))
```

```python
KNN1=KNeighborsClassifier()
start_time=time.time()
KNN1.fit(x_train,y_train)
end_time=time.time()
KNNt=end_time-start_time
KNN1_pred=KNN1.predict(x_test)
print("Accuracy of KNN:",accuracy_score(KNN1_pred,y_test))
print(classification_report(KNN1_pred,y_test))
```

```python
GNB1=GaussianNB()
start_time=time.time()
GNB1.fit(x_train,y_train)
end_time=time.time()
GNBt=end_time-start_time
GNB1_pred=GNB1.predict(x_test)
print("Accuracy of GNB:",accuracy_score(GNB1_pred,y_test))
print(classification_report(GNB1_pred,y_test))
```

```python
svm1=SVC()
start_time=time.time()
svm1.fit(x_train,y_train)
end_time=time.time()
svmt=end_time-start_time
svm1_pred=svm1.predict(x_test)
print("Accuracy of SVM:",accuracy_score(svm1_pred,y_test))
print(classification_report(svm1_pred,y_test))
```

```python
etc1=ExtraTreesClassifier()
start_time=time.time()
etc1.fit(x_train,y_train)
end_time=time.time()
etct=end_time-start_time
etc1_pred=etc1.predict(x_test)
print("Accuracy of ETC:",accuracy_score(etc1_pred,y_test))
print(classification_report(etc1_pred,y_test))
```

```python
lgbm1 = LGBMClassifier()
start_time=time.time()
lgbm1.fit(x_train,y_train)
end_time=time.time()
lgbt=end_time-start_time
lgbm1_pred=lgbm1.predict(x_test)
print("Accuracy of LGB:",accuracy_score(lgbm1_pred,y_test))
print(classification_report(lgbm1_pred,y_test))
```

```python
xgb1 = XGBClassifier()
start_time=time.time()
xgb1.fit(x_train,y_train)
end_time=time.time()
xgbt=end_time-start_time
xgb1_pred=xgb1.predict(x_test)
print("Accuracy of LGB:",accuracy_score(xgb1_pred,y_test))
print(classification_report(xgb1_pred,y_test))
```

```python
def preprocess_and_test(rf1,random_record_number):
    testing =pd.read_csv(r"C:\Vs code\Obisity-Level-Detection\ObesityDataSet.csv")
    le=LabelEncoder()
    for i in cat_col:
        testing[i]=le.fit_transform(testing[i])
    testing=pd.DataFrame(testing.iloc[random_record_number]).T
    class_value = testing['NObeyesdad'].iloc[0]
    print("Actual Value is",end=' ')
    if class_value == 0.0:
        print("Insufficient_Weight")
    elif class_value == 1.0:
        print("Normal_Weight")
    elif class_value == 2.0:
        print("Obesity_Type_I")
    elif class_value == 3.0:
        print("Obesity_Type_II")
    elif class_value == 4.0:
        print("Obesity_Type_III")
    elif class_value == 5.0:
        print("Overweight_Level_I")
    else:
        print("Overweight_Level_II")
    testing.drop(columns=['NObeyesdad'],inplace=True)
    
    print("Predicted is ",end='')
    if lgbm1.predict(testing) == 0.0:
        print("Insufficient_Weight")
    elif lgbm1.predict(testing) == 1.0:
        print("Normal_Weight")
    elif lgbm1.predict(testing) == 2.0:
        print("Obesity_Type_I")
    elif lgbm1.predict(testing) == 3.0:
        print("Obesity_Type_II")
    elif lgbm1.predict(testing) == 4.0:
        print("Obesity_Type_III")
    elif lgbm1.predict(testing) == 5.0:
        print("Overweight_Level_I")
    else:
        print("Overweight_Level_II") 

preprocess_and_test(rf1,75)    
```
